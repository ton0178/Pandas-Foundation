```python
import pandas as pd
```

```python
# create dataframe from scratch
raw_data = {
    "name":["Toy","Joe","Mary","John","Anna"],
    "age:":[33,25,20,22,31],
    "gender":["M","M","F","M","F"]
}
df = pd.DataFrame(raw_data)

df
```

```python
df['city'] = ['London','London','London','Manchester','Liverpool']
df
```

```python
df.shape
```

```python
# drop column city
df = df.drop('city',axis=1)
```

```python
df
```

```python
# remove index = 2
df = df.drop(2,axis = 0)
```

```python
df
```

```python
# reset index
df = df.reset_index(drop = True)
df
```

```python
# column names
list(df.columns)
```

```python
# rename column
df.columns = ['nickname','age','sex']
df
```

```python
type(df['nickname'])
```

```python
type(df)
```

```python
# create a new series
s1 = pd.Series(['Mary',20,'F'],index=['nickname','age','sex'])
print(s1)
print(type(s1))
```

```python
# append
df = df.append(s1,ignore_index=True)
df
```

```python
s2 = pd.Series(['London','London','London','Manchester','Liverpool'])
df['city'] = s2
df
```

```python
# write csv file
df.to_csv('mydata.csv')
```
